```python
import pandas as pd
import numpy as np
import json

stanford = pd.read_json("out/stanford_grades.json")
stanford.head()
```

```python
stanfordGrades = {
    'A+': 4.3,
    'A': 4.0,
    'A-': 3.7,
    'B+': 3.3,
    'B': 3.0,
    'B-': 2.7,
    'C+': 2.3,
    'C': 2.0,
    'C-': 1.7,
    'D+': 1.3,
    'D': 1.0,
    'D-': 0.7,
    'F': 0
}
```

```python
stanfordAverageGrades = dict()
```

```python
for course in stanford.index:
    stanfordAverageGrades[course] = sum([stanfordGrades[key]*value for (key, value) in stanford.loc[course]['classes'].items()])
```

```python
allAverageGrades = np.array(list(stanfordAverageGrades.values()))
```

```python
stanfordDifficultyRatings = dict()
for key, value in stanfordAverageGrades.items():
    stanfordDifficultyRatings[key] = float(10 * np.mean(allAverageGrades > value))
```

```python
with open(f"out/stanford_ratings.json", 'w') as f:
    json.dump(stanfordDifficultyRatings, f, indent=2)
```
