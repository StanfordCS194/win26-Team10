```python
from PyPDF2 import PdfReader
import re
```

```python
pattern = r"\b\d{4}-\d{4}\s+(?:Autumn|Winter|Spring|Summer)\b"

test_filename = "../transcripts/niall.pdf"
reader = PdfReader(test_filename)

def split(txt):
    items = []
    matches = re.finditer(pattern, txt)
    locations = [(match.start(), match.end()) for match in matches]

    for i in range(len(locations)):
        start = locations[i][0]
        if i == len(locations) - 1:
            item = txt[start:]
        else:
            end = locations[i+1][0]
            item = txt[start:end]

        items.append(item)
            
    return items

items = []

for i in range(0, len(reader.pages)):
    pageText = reader.pages[i].extract_text()
    items += split(pageText)

print(items)
```

```python
for item in items:
    print(item)
    print()
    print("__________")
```
